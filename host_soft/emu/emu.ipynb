```python
from math import sqrt
from valurap.asg import Asg, ProfileSegment, PathSegment

class ApgState(object):
    def __init__(self):
        self.x = 0
        self.v = 0
        self.a = 0
        self.j = 0
        self.jj = 0
        self.target_v = 0
        self.target_v_set = False
        
    def load(self, seg):
        if seg.x is not None:
            self.x = seg.x * 2**32
        if seg.v is not None:
            self.v = seg.v
        if seg.a is not None:
            self.a = seg.a
        if seg.j is not None:
            self.j = seg.j
        if seg.jj is not None:
            self.jj = seg.jj
        if seg.target_v is not None:
            self.target_v = seg.target_v
            self.target_v_set = True
        else:
            self.target_v_set = False
            
    def step(self):
        next_x = self.x + self.v * 50000
        next_v = round(self.v + (self.a / 256))
        next_a = self.a + self.j
        next_j = self.j + self.jj
        next_jj = self.jj
        if self.target_v_set:
            if self.v == self.target_v:
                next_jj = 0
                next_j = 0
                next_a = 0
                next_v = self.target_v
            elif ((self.v < self.target_v and next_v > self.target_v) or 
                  (self.v > self.target_v and next_v < self.target_v)):
                next_jj = 0
                next_j = 0
                next_v = self.target_v
                next_a = (self.target_v - self.v) * 256
             
        #effective_v = (self.v + next_v) / 2
        effective_v = next_v

        self.x = next_x
        self.v = next_v
        self.a = next_a
        self.j = next_j
        self.jj = next_jj
        
    
def emulate(profile, verbose=False):
    apg_states = {
        "X": ApgState(),
        "Y": ApgState(),
        "Z": ApgState()
    }
    ts = 0
    steps = {}
    for dt, segs in profile:
        print(dt)
        ts_start = ts
        ts += dt
        for seg in segs:
            print("  ",seg)
            state = apg_states[seg.apg.name]
            state.load(seg)
            prefix = seg.apg.name + "_"
            
            if not verbose:
                print("    {:6d} {:10.3f} {:10.1f} {:10.1f}".format(ts_start, state.x / 2.0**32, state.v, state.a))
                print("                 ...")

            last_v = None
            first_v = 0
            for i in range(dt):
                step_data = steps.setdefault(ts_start + i, {'ts': ts_start + i})
                step_data[prefix + "jj"] = state.jj
                step_data[prefix + "j"] = state.j
                step_data[prefix + "a"] = state.a
                step_data[prefix + "v"] = state.v
                step_data[prefix + "x"] = state.x

                if state.v != last_v:
                    if verbose:
                        if first_v > 0:
                            if first_v > 1:
                                print("        ... {} ...".format(first_v - 1))
                            print("    {:6d} {:10.3f} {:10.1f} {:10.1f}".format(ts_start + i - 1, prev_x / 2.0**32, prev_v, prev_a))
                        print("    {:6d} {:10.3f} {:10.1f} {:10.1f}".format(ts_start + i, state.x / 2.0**32, state.v, state.a))
                    last_v = state.v
                    first_v = 0
                else:
                    if not verbose and first_v == 0:
                        print("    {:6d} {:10.3f} {:10.1f} {:10.1f}".format(ts_start + i - 1, prev_x / 2.0**32, prev_v, prev_a))

                    first_v += 1
                    
                prev_x = state.x
                prev_v = state.v
                prev_a = state.a
                state.step()

            if verbose:
                if first_v != 0:
                    if first_v > 1:
                        print("           ... {} ...".format(first_v - 1))
                    print("    {:6d} {:10.3f} {:10.1f} {:10.1f}".format(ts_start + i, state.x / 2.0**32, state.v, state.a))
            else:
                if first_v > 1:
                    print("                 ...")

                print("    {:6d} {:10.3f} {:10.1f} {:10.1f}".format(ts_start + i, state.x / 2.0**32, state.v, state.a))
            

    steps = [a[1] for a in sorted(steps.items())]
    return steps
                
    
p.axes["X1"].apg = p.apgs["X"]
p.axes["X2"].apg = p.apgs["Y"]
p.axes["Y"].apg = p.apgs["Z"]

pr = p.plan_move(Y=10000)

%matplotlib inline

import pandas as pd
steps = emulate(pr, True)
steps = pd.DataFrame(steps)
steps.plot(x = "ts", y = ["Z_a"])

```

```python
from ortools.sat.python import cp_model
import time

def solve_model(target_x, target_v, max_a = 20000):
    model = cp_model.CpModel()

    # optimized profile parameters
    t = model.NewIntVar(1, 100000, 't')
    j = model.NewIntVar(1, 1000000, 'j')
    jj = model.NewIntVar(-100000, -1, 'jj')

    # errors
    dx = model.NewIntVar(0, int(2**32 * 256 / 50000 / 2), 'dx') # X error - +- 0.5 step
    dv = model.NewIntVar(0, 100000*256, 'dv') # V error +- 14mm/s
    dj = model.NewIntVar(0, 20, 'dj') # residual J
    
    # end of move X
    # x = j*t*(t-1)*(t-2)/6 + jj*t*(t-1)*(t-2)*(t-3)/24
    # adding x24 = x * 24 and rounding error x24_delta to keep it integer
    x = model.NewIntVar(1, 2**60, 'x')
    x24 = model.NewIntVar(-2**60, 2**60, 'x24')
    x24_delta = model.NewIntVar(0, 23, 'x24_delta')

    # end of move V
    # v = j*t*(t-1)/2 + jj*t*(t-1)*(t-2)/6
    # adding v6 = v * 6 and rounding error v6_delta to keep it integer
    v = model.NewIntVar(1, 2**60, 'v')
    v6 = model.NewIntVar(1, 2**60, 'v')
    v6_delta = model.NewIntVar(0, 5, 'v6_delta')
    
    # declare polynominal members
    jt = model.NewIntVar(1, 2**50, 'jt')
    jt2 = model.NewIntVar(1, 2**50, 'jt2')
    jt3 = model.NewIntVar(1, 2**50, 'jt3')
    
    jjt = model.NewIntVar(-2**50, -1, 'jjt')
    jjt2 = model.NewIntVar(-2**50, -1, 'jjt2')
    jjt3 = model.NewIntVar(-2**50, -1, 'jjt3')
    jjt4 = model.NewIntVar(-2**50, -1, 'jjt4')

    model.AddProdEquality(jt, [j, t])
    model.AddProdEquality(jt2, [jt, t])
    model.AddProdEquality(jt3, [jt2, t])

    model.AddProdEquality(jjt, [jj, t])
    model.AddProdEquality(jjt2, [jjt, t])
    model.AddProdEquality(jjt3, [jjt2, t])
    model.AddProdEquality(jjt4, [jjt3, t])
    
    # X discrete integral
    model.Add(x24 == jjt4 - 6 * jjt3 + 4 * jt3 + 11*jjt2 - 12*jt2 - 6 * jjt + 8 * jt + x24_delta)
    model.AddProdEquality(x24, [x, 24])

    # V discrete integral
    model.Add(v6 == jjt3 - 3 * jjt2 + 3 * jt2 + 2*jjt - 3*jt + v6_delta)
    model.AddProdEquality(v6, [v, 6])
              
    
    
    # target X
    model.Add((x - int(target_x * 2**32 * 256 / 50000) == dx) or (x - int(target_x * 2**32 * 256 / 50000) == -dx))
    # target V
    model.Add((v - int(target_v * 256) == dv) or (v - int(target_v * 256) == -dv))
    # Accel limit
    model.Add(jt < max_a * 256 * 2)
    
    # accel to 0 at the end of move
    model.Add((jjt + 2 * j == dj) or (jjt + 2 * j == -dj))
    
    # minimize errors
    model.Minimize(dv)
    model.Minimize(dx)
    model.Minimize(dj)
    
    # minimize total time
    model.Minimize(t)

    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    print(status)
    if True or status == cp_model.FEASIBLE:
        print('t = %i' % solver.Value(t))
        print('j = %i' % solver.Value(j))
        print('dj = %i' % solver.Value(dj))
        print('jj = %i' % solver.Value(jj))
        print('x = %.2f' % (solver.Value(x)/(2**32)/256*50000))
        print('x24_delta = %i' % solver.Value(x24_delta))
        print('dx = %.2f' % (solver.Value(dx)/(2**32)/256*50000))
        print('x24 = %i' % solver.Value(x24))
        print('v = %.2f' % (solver.Value(v)/256))
        print('dv = %.2f' % (solver.Value(dv)/256))
    return {
        "t": solver.Value(t),
        "j": solver.Value(j),
        "jj": solver.Value(jj),
    }

t = time.time()
solution = solve_model(1000, 500000)
t2 = time.time()
print("solving time", (t2-t)*1000)

apg = p.apgs["X"]

pr = [
    [solution["t"], [ProfileSegment(apg=apg, j=solution["j"], jj=solution["jj"], target_v = None)]],
]
steps = emulate(pr, False)
steps = pd.DataFrame(steps)
steps.plot(x = "ts", y = ["X_a"])
steps.plot(x = "ts", y = ["X_v"])
```
